```python
df1 = pd.read_csv('/content/ordenada.txt', sep = ',')
df2 = pd.read_csv('/content/inversa.txt', sep = ',')
df3 = pd.read_csv('/content/aleatoria.txt', sep = ',')
```

```python
x = df1['tamanho']
sel_1 = df1['selection']
ins_1 = df1['insertion']
bub_1 = df1['bubble']
mer_1 = df1['merge']
she_1 = df1['shell']
qui_1 = df1['quick']
hea_1 = df1['heap']

sel_2 = df2['selection']
ins_2 = df2['insertion']
bub_2 = df2['bubble']
mer_2 = df2['merge']
she_2 = df2['shell']
qui_2 = df2['quick']
hea_2 = df2['heap']

sel_3 = df3['selection']
ins_3 = df3['insertion']
bub_3 = df3['bubble']
mer_3 = df3['merge']
she_3 = df3['shell']
qui_3 = df3['quick']
hea_3 = df3['heap']
```

```python
plt.figure(figsize=(8,8))
plt.plot(x,sel_1, color='lawngreen', label='Selection')
plt.plot(x,ins_1, color='orange', label='Insertion')
plt.plot(x,bub_1, color='red', label='Bubble')
plt.plot(x,mer_1, color='royalblue', label='Merge sort')
plt.plot(x,she_1, color='mediumblue', label='Shell sort')
plt.plot(x,qui_1, color='green', label='Quick sort')
plt.plot(x,hea_1, color='indigo', label='Heap sort')
plt.xlim(min(x), max(x))
plt.title('Sequências ordenadas', fontweight='bold')
plt.xlabel('Sequências numéricas de 1 a n', fontstyle='italic')
plt.ylabel('Tempo em segundos', fontstyle='italic')
plt.legend()
plt.savefig('geral_ordenada.png',dpi=600)
plt.show()
```
